```python
def validate_username(username):
    """
    ユーザー名をバリデーションする関数
    
    Args:
        username (str): 検証するユーザー名
    
    Returns:
        bool: バリデーション結果（True: 有効, False: 無効）
    """
    # 文字数のチェック（5文字以上15文字以内）
    if len(username) < 5 or len(username) > 15:
        print("エラー: ユーザー名は5文字以上15文字以内で入力してください")
        return False
    
    # 空白のチェック
    if " " in username:
        print("エラー: ユーザー名に空白を含めることはできません")
        return False
    
    # 英数字のみかチェック
    if not username.isalnum():
        print("エラー: ユーザー名は英数字のみ使用できます")
        return False
    
    # すべての条件を満たす場合
    return True


# ユーザーにユーザー名を入力させる
username = input("ユーザー名を入力してください: ")

# バリデーション関数を呼び出して判定
if validate_username(username):
    print("ユーザー名は有効です")
else:
    print("不正なユーザー名です")
```
