```python
import urllib.request
import datetime


def save_url_content(url):
    """
    指定したURLのコンテンツを取得し、現在の日時をファイル名にして保存する
    
    Args:
        url (str): 取得するWebページのURL
    
    Returns:
        str: 保存したファイル名
    """
    # URLからデータを取得
    response = urllib.request.urlopen(url)
    
    # HTMLデータを読み込み、UTF-8文字列にデコードして先頭500文字を取得
    html = response.read().decode("utf-8")
    content = html[:500]
    
    # 現在の日時からファイル名を生成（例: 20241030_155221.txt）
    now = datetime.datetime.now()
    file_name = now.strftime("%Y%m%d_%H%M%S.txt")
    
    # ファイルに保存
    with open(file_name, "w", encoding="utf-8") as f:
        f.write(content)
    
    # ファイル名を返す
    return file_name


# ユーザーにURLを入力させる
url = input("URLを入力してください: ")

# 関数を呼び出してファイルに保存
file_name = save_url_content(url)

# 保存したファイル名を表示
print(f"Content saved in: {file_name}")
```
