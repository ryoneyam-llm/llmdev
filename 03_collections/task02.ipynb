# 課題2: 単語の文字数カウント

英単語を入力し、ABC順に並べ替えて表示します。
また、すべての英単語に含まれるアルファベットごとの数も表示します。

```python
# 英単語を格納するための空のリスト
words = []

# アルファベットごとの数を格納する辞書
letter_count = {
    'a': 0, 'b': 0, 'c': 0, 'd': 0, 'e': 0, 'f': 0, 'g': 0, 'h': 0,
    'i': 0, 'j': 0, 'k': 0, 'l': 0, 'm': 0, 'n': 0, 'o': 0, 'p': 0,
    'q': 0, 'r': 0, 's': 0, 't': 0, 'u': 0, 'v': 0, 'w': 0, 'x': 0,
    'y': 0, 'z': 0
}

# 英単語の入力を繰り返し受け付ける
while True:
    word = input("英単語を入力してください：")
    
    # 空文字列が入力されたら終了
    if word == "":
        break
    
    # 英単語をリストに追加
    words.append(word)
    
    # 英単語の各文字をカウント
    for char in word.lower():
        if char in letter_count:
            letter_count[char] += 1

# 英単語をABC順に並べ替え
words.sort()

# 英単語のリストを表示
print(f"入力した英単語：{words}")

# アルファベットごとの数を表示（0より大きい文字のみ）
for letter, count in letter_count.items():
    if count > 0:
        print(f"{letter}が{count}個ありました")
```
