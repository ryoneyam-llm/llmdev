# 課題4: 数値計算の例外処理プログラム

2つの数値を入力して割り算を行います。
- 文字が入力された場合: 「エラー: 数値を入力してください」
- 0で割る場合: 「エラー: 0で割ることはできません」
- 正常に計算できた場合: 計算結果を表示

```python
try:
    # 2つの数値を入力
    num1 = float(input("1つ目の数を入力してください: "))
    num2 = float(input("2つ目の数を入力してください: "))
    
    # 割り算を実行
    result = num1 / num2
    
    # 結果を表示
    print(f"計算結果: {result}")
    
except ValueError:
    # 数値以外が入力された場合
    print("エラー: 数値を入力してください")
    
except ZeroDivisionError:
    # 0で割ろうとした場合
    print("エラー: 0で割ることはできません")
```
