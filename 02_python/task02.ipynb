# 課題2: 送料込価格計算プログラム

税抜価格を入力し、消費税込み価格と送料を合計した送料込価格を計算します。
- 税込価格が2000円以上の場合は送料無料
- 税込価格が2000円未満の場合は送料350円
- 消費税率は10%

```python
# 税抜価格を入力
price_before_tax = int(input("税抜価格を入力してください: "))

# 消費税率
tax_rate = 0.1

# 税込価格を計算
price_with_tax = int(price_before_tax * (1 + tax_rate))

# 送料を判定
if price_with_tax >= 2000:
    shipping_fee = 0
    print("送料は無料です")
else:
    shipping_fee = 350
    print(f"送料は{shipping_fee}円です")

# 送料込税込価格を計算
total_price = price_with_tax + shipping_fee

# 結果を表示
print(f"送料込税込価格は{total_price}円です")
```
