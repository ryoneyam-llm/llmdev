# 課題3: 10年後の金額を複利で予測するプログラム

元金と年利を入力し、10年後の金額を複利で計算します。

複利とは、利子を次期の元金に組み入れる方式です。
例: 元金1000円、年利5%の場合
- 1年後: 1000 * (1 + (5 / 100)) = 1050.0
- 2年後: 1050.0 * (1 + (5 / 100)) = 1102.5
- 3年後: 1102.5 * (1 + (5 / 100)) = 1157.625

```python
# 元金を入力
principal = float(input("元金を入力してください: "))

# 年利を入力
interest_rate = float(input("年利を入力してください: "))

# 現在の金額を元金で初期化
amount = principal

# 10年間の複利計算
for year in range(1, 11):
    amount = amount * (1 + (interest_rate / 100))

# 10年後の金額を表示（小数点以下2桁まで）
print(f"10年後の金額: {amount:.2f}")
```
